# Build Downstream

```python
def find_children(data, source, network_role=None):
    rows = data[data.source == source]
    
    if rows.shape[0] == 0:
        return 
    
    nest = [source]
    for i, row in rows.iterrows():
        r = []
        target = row.source
        next_target = row.target
            
        if data[data.source == next_target].shape[0] == 0:
            r.append([next_target])
        else:
            r.append(find_children(data, next_target))
            
        nest.extend(r)
    
    return nest
```

```python
def build_edges(children):
    edges = []
    parent_node = children[0]
    branches = children[1:]
    
    for b in branches:
        edges.append((parent_node, b[0]))
        
        if type(b) == list and len(b) == 1:
            edges.append((parent_node, b[0]))
        else:
            nested_edges = build_edges(b)
            edges.extend(nested_edges)
    
    seen = set()
    edges =  [x for x in edges if not (x in seen or seen.add(x))]
    return edges
```

```python
def build_edge_data(data, paths):
    edge_num = 0
    columns = ['source', 'cumulative_volume_af', 'transmission_kwh/af', 
               'treatment_kwh/af','used_vol_af']
    
    edge_dicts = []
    for i in paths:
        rows = data[(data.source == i[0]) & (data.target == i[1])]
        
        edge_dict = rows[columns].to_dict('records')
        
        edge_dict[0]['number'] = edge_num
        edge_dict[0]['target'] = i[1]
        
        edge_num += 1
        
        edge_dicts.append(edge_dict)
    
    edge_dicts = [i for sub in edge_dicts for i in sub]
    
    return edge_dicts
```

```python
def build_graph(data, source):
    nodes = {}
    edges = {}
    node_num = 0
    
    graph = nx.MultiDiGraph()
    
    children = find_children(data, source, [])
    
    columns = ['source', 'cumulative_volume_af', 'transmission_kwh/af', 
               'treatment_kwh/af','used_vol_af']
    
    paths = build_edges(children)
    
    edge_dicts = build_edge_data(data, paths)
    
    for i in range(len(paths)):
        edge = paths[i]
        edges[(edge[0], edge[1])] = edge_dicts[i]
        
        if edge[0] not in nodes.keys():
            nodes[edge[0]] = node_num
            node_num += 1
        
        if edge[1] not in nodes.keys():
            nodes[edge[1]] = node_num
            node_num += 1
    
    graph.add_nodes_from(nodes)
    graph.add_edges_from(edges)
    
    nx.draw_networkx(graph)
    
    print(nx.nx_pydot.to_pydot(graph))
    
    return nx.drawing.nx_pydot.write_dot(graph, 'graph.dot')
```

```python
build_graph(ten, 'SW_CVP-Coyot')
```

# Build Upstream

```python
def get_parents(data, target):
    rows = data[data.target == target]
    
    if rows.shape[0] == 0:
        return 
    
    child = [target]
    for i, row in rows.iterrows():
        p = []
        parent = row.target
        next_target = row.source
        if data[data.target == next_target].shape[0] == 0:
            p.append([next_target])
        else:
            p.append(get_parents(data, next_target))
        
        child.extend(p)
    
    return child
```

```python
def build_parents(parents):
    edges = []
    child_node = parents[0]
    parents_above = parents[1:]
    
    for p in parents_above:
        edges.append((p[0], child_node))
        
        if type(p) == list and len(p) == 1:
            edges.append((p[0], child_node))
        else:
            nested_edges = build_parents(p)
            edges.extend(nested_edges)
    
    seen = set()
    edges =  [x for x in edges if not (x in seen or seen.add(x))][::-1]
    return edges
```

```python
def graph_edges(path, down=True):
    edges = []
    first_node = path[0]
    branches = path[1 :]
    
    if down:
        for b in branches:
            edges.append((first_node, b[0]))
        
            if type(b) == list and len(b) == 1:
                edges.append((first_node, b[0]))
            else:
                nested_edges = graph_edges(b, down)
                edges.extend(nested_edges)

        seen = set()
        edges =  [x for x in edges if not (x in seen or seen.add(x))]
        return edges
    
    else:
        for b in branches:
            edges.append((b[0], first_node))

            if type(b) == list and len(b) == 1:
                edges.append((b[0], first_node))
            else:
                nested_edges = graph_edges(b, down)
                edges.extend(nested_edges)

        seen = set()
        edges =  [x for x in edges if not (x in seen or seen.add(x))][::-1]
        return edges
```

```python
def build_graph_upstream(data, source):
    nodes = {}
    edges = {}
    node_num = 0
    
    graph = nx.MultiDiGraph()
    
    children = get_parents(data, source)
    
    columns = ['source', 'cumulative_volume_af', 'transmission_kwh/af', 
               'treatment_kwh/af','used_vol_af']
    
    paths = graph_edges(children, down=False)
    
    edge_dicts = build_edge_data(data, paths)
    
    for i in range(len(paths)):
        edge = paths[i]
        edges[(edge[0], edge[1])] = edge_dicts[i]
        
        if edge[0] not in nodes.keys():
            nodes[edge[0]] = node_num
            node_num += 1
        
        if edge[1] not in nodes.keys():
            nodes[edge[1]] = node_num
            node_num += 1
    
    graph.add_nodes_from(nodes)
    graph.add_edges_from(edges)
    
    nx.draw_networkx(graph)
    
    return edges
```
