```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import pandas as pd
import json
```

```python
ucdd = pd.ExcelFile('data/Utility codes data dictionary.xlsx')
```

```python
basin = ucdd.parse(2)
basin.head()
```

```python
{
    'Basins':{
        'ALEXANDER VALLEY':{
            'Subbasin_Name': 'CLOVERDALE AREA',
            'Basin_code': 'GW1801-1-54',
            'Subbasin_code': 'GW1801-1-54.02'
        }
    }
}
```

```python
dd = {'basin_name': {
        'ALEXANDER VALLEY':{
            'subbasin_name': 'CLOVERDALE AREA',
            'basin_code': 'GW1801-1-54',
            'subbasin_code': 'GW1801-1-54.02'}}}

for index in range(1, basin.shape[0]):
    row_data = basin.loc[index].values
      
    if type(row_data[1]) == str:
        sub_name = row_data[1]
    else:
        sub_name = ''
      
    if type(row_data[2]) == str:
        basin_code = row_data[2]
    else:
        basin_code = ''
      
    if type(row_data[3]) == str:
        sb_code = row_data[3]
    else:
        sb_code = ''
      
    dd['basin_name'][row_data[0]] = {'subbasin_name':sub_name,
                                     'basin_code':basin_code,
                                     'subbasin_code':sb_code}
```

```python
print(json.dumps(dd, indent=4))
```

```python
imp_surf = ucdd.parse(1)
nodes = imp_surf.iloc[:, :2]
node_acronyms = imp_surf[['Acronym', 'System']].iloc[0:6]
```

```python
nodes.head()
```

```python
node_dict = {}
for row in nodes.iterrows():
    node_dict[row[1][0]] = {'node_def':str(row[1][1])}
```

```python
dd['node_name'] = node_dict
```

```python
node_acronyms
```

```python
node_acro = {}
for row in node_acronyms.itertuples():
    node_acro[row[1]] = {'system': row[2]}
print(json.dumps(node_acro, indent=4))
```

```python
dd['node_acronyms'] = node_acro
```

```python
utilities = ucdd.parse(0)
utilities.head()
```

```python
util = pd.read_csv('data/Utility codes data dictionary - utilities.csv')
```

```python
utility_dict={}
for row in util.itertuples(index=False):
    util_code = row[1]
    utility_dict[util_code] = {'supplier': row[0],
                            'wholesale/retail': row[2]}
print(json.dumps(utility_dict, indent=2))
```

```python
dd['utility'] = utility_dict
```

```python
with open('data_dict_utility.json', 'w') as f:
    f.write(json.dumps(dd, indent=4))
```
